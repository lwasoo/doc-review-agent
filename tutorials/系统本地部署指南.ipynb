### 1.1 操作系统
- **Linux**: Ubuntu 20.04+ / CentOS 7+ / Debian 10+（推荐）

## 📦 四、安装项目依赖

### 4.2 运行安装脚本

````python
!./install.sh
````

**安装过程说明**：

脚本会自动执行以下步骤：

#### 1. 环境检查
- 检测 Node.js、npm、Python 版本
- 验证必要工具是否就绪

#### 2. 后端依赖安装
```
✅ 创建 Python 虚拟环境 (venv)
✅ 升级 pip 到最新版本
✅ 安装 requirements.txt 中的所有依赖
```

主要依赖包括：
- `fastapi` - 后端 API 框架
- `langchain==1.1.3` - LLM 编排框架
- `langchain-deepseek==1.0.1` - DeepSeek 集成
- `langchain-openai==1.1.3` - OpenAI 集成
- `pydantic==2.11.1` - 数据验证
- `aiosqlite==0.20.0` - 异步数据库
- `pysqlite3-binary==0.5.4.post2` - SQLite3 兼容层
- `uvicorn[standard]==0.32.0` - ASGI 服务器
- `httpx==0.28.1` - HTTP 客户端

#### 3. 前端依赖安装
```
✅ 安装 npm 依赖包
```

主要依赖包括：
- `react` - 前端框架
- `@fluentui/react-components` - Microsoft Fluent UI 组件库
- `vite` - 构建工具
- `react-pdf` - PDF 查看器
- `react-router-dom` - 路由管理

#### 4. 自动创建配置文件
- 如果不存在 `.env` 文件，会从 `.env.tpl` 模板复制

**⏱️ 预计耗时**: 5-10 分钟（取决于网络速度）

### 4.3 常见安装问题


## 🔑 五、配置环境变量

### 5.1 编辑配置文件

````python
# 使用你喜欢的编辑器
!nano app/api/.env
# 或
# !vim app/api/.env
# 或
# !code app/api/.env  # VS Code
````

### 5.2 必填配置项

```ini

# DeepSeek API（推荐）

# OPENAI_API_KEY=sk-xxxxxxxxxxxxxxxxxxxxxxxx
# OPENAI_BASE_URL=https://api.openai.com/v1

DEFAULT_MODEL_NAME=deepseek-chat

# ========== 数据库配置 ==========

# ========== 服务端口 ==========
# API_PORT=8000  # 默认 8000，一般无需修改

# ========== 日志级别 ==========
LOG_LEVEL=INFO  # 可选: DEBUG, INFO, WARNING, ERROR
```

## 🐛 九、故障排查

### 9.1 后端无法启动

#### 症状: 8000 端口被占用

```bash
# 查找占用端口的进程
lsof -i :8000
# 或
netstat -tunlp | grep 8000

# 杀死进程
kill -9 <PID>
```

#### 症状: 数据库连接失败

```bash
# 检查数据库文件权限
ls -la app/api/data/

# 重新初始化数据库
rm -rf app/api/data/review.db
# 重启服务会自动创建
```

#### 症状: ModuleNotFoundError

```bash
# 确认虚拟环境已激活
cd app/api
source venv/bin/activate
which python  # 应该指向 venv 中的 python

# 重新安装依赖
pip install -r requirements.txt
```

### 9.2 前端无法连接后端

#### 症状: API 请求 404 或超时

**解决方案**:

1. 检查后端是否正常运行
   ```bash
   curl http://localhost:8000/api/v1/health
   ```

2. 检查 Vite 代理配置
   ```bash
   # 确认 app/ui/vite.config.ts 中的代理配置
   # target 应为 http://127.0.0.1:8000
   cat app/ui/vite.config.ts | grep target
   ```

3. 清除浏览器缓存并刷新
   - Chrome: `Ctrl+Shift+Delete`
   - 或使用无痕模式测试

### 9.3 虚拟环境问题

#### 症状: 找不到 Python 模块

```bash
# 重新创建虚拟环境
rm -rf app/api/venv
cd app/api
python -m venv venv
source venv/bin/activate
pip install -r requirements.txt
```

### 9.4 依赖版本冲突

```bash
# 查看已安装的包
pip list

# 查看冲突信息
pip check

# 升级特定包
pip install --upgrade <package-name>

# 完全重装
rm -rf app/api/venv
./install.sh
```

### 9.5 前端构建失败

```bash
# 清理 node_modules
cd app/ui
rm -rf node_modules package-lock.json
npm install

# 或使用 npm ci（更快更可靠）
npm ci
```

### 9.6 权限问题

```bash
# 修复文件所有权
sudo chown -R $USER:$USER /path/to/ai-document-review

# 修复执行权限
chmod +x install.sh start.sh stop.sh
chmod +x app/ui/node_modules/.bin/*
```

### 9.7 日志查看

```bash
# 查看后端日志
tail -f app/api/logs/app.log

# 查看系统日志
journalctl -u ai-document-review -f
```
